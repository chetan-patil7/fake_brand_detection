```python
# import necessary libraries
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
from tensorflow.keras.models import load_model

# Load the model
model = load_model('D:/streamlit/cnn_model_final.h5')
```

```python
from PIL import Image
# Define the image path
image_file = 'D:/Final_year_project/Brands/Electronics/HP/Original/hp (10)'

# Load the image
image = Image.open(image_file)

# Convert the image to grayscale
image = image.convert("L")

# Preprocess the image
image = image.resize((256, 256))  # Resize the image to 256x256
image_array = np.array(image)  # Convert image to numpy array
image_array = np.expand_dims(image_array, axis=0)  # Add extra dimension to match model input shape

# Normalize the input
image_array = image_array / 255
```

```python
import matplotlib.pyplot as plt

# Predict the label
predictions = model.predict(image_array)
predicted_label = np.argmax(predictions)

# Define the class labels
classes = ["Original", "Fake"]

# Get the actual label from the image path
actual_label = "Original" if "Original" in image_file else "Fake"

# Display the image and predicted label
plt.imshow(image, cmap="gray")
plt.title(f"Actual: {actual_label} / Predicted: {classes[predicted_label]}")
plt.axis("off")
plt.show()
```
